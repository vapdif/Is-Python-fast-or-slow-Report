### Generating 1-D and 2-D Arrays with Dimension 9

```python
# Generate 1D array

# Initialize an empty list
array_1d_9 = []

# Use a for loop to add elements to the array
for i in range(9):  # This will generate numbers from 0 to 8
    array_1d_9.append(i)

# Display the 1D array
print(array_1d_9)
```

```python
# Define the number of rows and columns for our 2D array
num_rows = 3
num_columns = 9

# Initialize the 2D array as an empty list
array_2d_9 = []

# Use a for loop to generate each row of the 2D array
for i in range(num_rows):
    # Initialize the current row as an empty list
    row = []
    # Add elements to the current row
    for j in range(num_columns):
        # Calculate the current element's value based on the row and column index
        element = i * num_columns + j
        # Add the element to the current row
        row.append(element)
    # Add the complete row to the 2D array
    array_2d_9.append(row)

# Display the generated 2D array
for row in array_2d_9:
    print(row)
```

```python
# Perform the multiplication of the 2D array by the 1D array iteratively

# Start time measurement
start_time = time.perf_counter()



# Initialize a list to store the result
iterative_result = []

# Iterate over each row of the 2D array
for row in array_2d_9:
    # Initialize the sum for this row
    row_sum = 0
    # Iterate over each element of the row and the 1D array simultaneously
    for row_element, array_1d_element in zip(row, array_1d_9):
        # Multiply the elements and add them to the row's total
        row_sum += row_element * array_1d_element
    # Add the result of the row's sum to the final result
    iterative_result.append(row_sum)

# Display the iterative result
print(iterative_result)



# Calculate elapsed time
elapsed_time = time.perf_counter() - start_time

# Show runtime
print(f"\nTime is: {elapsed_time:.5f} seconds")
```
